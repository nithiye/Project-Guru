```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Read file with countries' GDP numbers
gdp_data_csv = "Resources/CountryGDP.csv"
# read a catalog of countries with their respective regions
country_cat_csv = "Resources/Region_country_catalog.csv"

# Read data into Pandas data frames
gdp_data_df = pd.read_csv(gdp_data_csv)
country_cat_df = pd.read_csv(country_cat_csv)
```

```python
# join the GDP data to the catalog to get rid of all rows that DO NOT contain country information
gdp_total_df = pd.merge(country_cat_df[["Country"]],gdp_data_df, on="Country")
#gdp_total_df.loc[gdp_total_df["Country"] == "India"].dtypes
```

```python
# Top 20 migrant countries
top20_ctrys_lst = ['India', 'Mexico', 'Russian Federation', 'China', 'Bangladesh',
       'Ukraine', 'Pakistan', 'Philippines', 'Afghanistan', 'United Kingdom',
       'Poland', 'Germany', 'Kazakhstan', 'Syrian Arab Republic', 'Indonesia',
       'State of Palestine', 'Romania', 'Turkey', 'Italy', 'Egypt']
```

```python
# Filter top 20 migrant countries GDP dataframe 
# Possible indicators: Personal remittances, received (% of GDP), Personal remittances, received (current US$),
# GDP growth (annual %), GDP (current US$)
gdp_total_df.loc[(gdp_total_df.Country.isin(top20_ctrys_lst)) & \
                 (gdp_total_df["Indicator Name"] == "Personal remittances, received (% of GDP)")]
```

```python
# Read migration information
migrate_csv = "Output/Clean_Top20_migrant.csv"

# Read data into Pandas data frames
top_20_ctrys_df = pd.read_csv(migrate_csv)
```

```python
# Merge both sets
all_info_df = pd.merge(top_20_ctrys_df,gdp_total_df, left_on="Origin", right_on="Country").drop("Country",axis=1)
```

```python
# Write the new dataframe to a new CSV file
all_info_df.to_csv("Output/Data_for_plotting.csv", index=False, header=True)
```
