```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

country_data = "../Output/Data_for_plotting.csv"

mig_data = pd.read_csv(country_data)

mig_data2 = pd.DataFrame(mig_data)

mig_data2.head()
```

```python
mig_df = mig_data2.filter(['Origin','2005_x','2010_x','2015_x','2017_x','Indicator','2005_y','2010_y','2015_y','2017_y'], axis=1)
mig_df
```

```python
Cash_Total = mig_df.iloc[:,7:12].sum(axis=1)
print(Cash_Total)
```

```python
Migrant_Total = mig_data.iloc[:,3:7].sum(axis=1)
print(Migrant_Total)
```

```python
mig_df["Cash_Total"] = mig_df.iloc[:,6:10].sum(axis=1)
mig_df
```

```python
mig_df["Migrant_Total"] = mig_df.iloc[:,1:5].sum(axis=1)
mig_df
```

```python
origin_total = mig_df.groupby(by="Origin")["Migrant_Total"].mean()
origin_total
```

```python
Personal_R = mig_df.loc[mig_df["Indicator"] == "Personal remittances, received (current US$)"]
GDP = mig_df.loc[mig_df["Indicator"] =="GDP (current US$)"]
GDP
```

```python
newdf = pd.DataFrame(columns=['x','y','s'])
newdf['x'] = Personal_R['Cash_Total'].values
newdf['y'] = GDP['Cash_Total'].values
newdf['s'] = (np.log2(GDP['Migrant_Total'].values) - 20) * 18
newdf
```

```python
newdf.plot.scatter(x='x', y='y', s = newdf['s'].values)
```

```python
# Build the scatter plots for each city types

plt.scatter(x=newdf['x'],y=newdf['y'],s= newdf['s']*10,marker= "o",c=s, label="Migrants", alpha=0.5, linewidth=1)

plt.title("Cash Remittances for Migrant Data",{"fontname":"Arial","fontsize":18})
plt.xlabel("Total Amount of Personal Remittances")
plt.ylabel("Total Amount of GDP")


# Create a legend
plt.legend(loc="best")

# Incorporate a text label regarding circle size

#Show Figure
plt.show()

# Save Figure
plt.savefig("HomeworkSP.png")
```
